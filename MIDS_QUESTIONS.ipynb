```python
no_of_products=50
while(no_of_products!=0):
    if no_of_products<=150 and no_of_products>0:
        product_need=int(input("How many products you want ? "))
        print("You have bought ",product_need," products")
        no_of_products-=product_need
        if no_of_products>=0:
            print("Remaining Products are : ",no_of_products)
        else:
            print("Out of Stock !!, !!")
```

### QUESTION 5

```python
total_marks=300
sub1=eval(input("Enter Marks in Subject 1 : "))
sub2=eval(input("Enter Marks in Subject 2 : "))
sub3=eval(input("Enter Marks in Subject 3 :"))
sum=sub1+sub2+sub3
percent=(sum/total_marks)*100
if percent>=91 and percent<=100:
    print("your grade is A+")
elif percent>82 and percent<=90:
    print("your grade is A")
elif percent>=75 and percent<=82:
    print("your grade is B+")
elif percent>=65 and percent<=74:
    print("your grade is B")
elif percent>=60 and percent<=64:
    print("your grade is C+")
elif percent>=50 and percent<=59:
    print("your grade is C")
else:
    print("You are Fail , Sorry!!")

```
